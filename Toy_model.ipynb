```python
import torch
import gpytorch
from matplotlib import pyplot as plt
from gpytorch.likelihoods import GaussianLikelihood

import numpy as np

%matplotlib inline
%load_ext autoreload 
%autoreload 2
```

Define the objective Function:

```python
def objective(x, noise = 0.1):
    # transform x:
    x = (x)*4. - 2.
    return 2 * torch.sin(5 * x) * (1 - torch.tanh(x ** 2)) + torch.rand(1) * noise + 2.

plt.plot(torch.linspace(0., 1., 1000), objective(torch.linspace(0.,1.,1000)))
plt.title('True Toy Function')
```

Generate training and test data:

```python
x_train = torch.rand(10)
y_train = objective(x_train)

x_valid = torch.linspace(0,1,100)
y_valid = objective(x_valid)
```
